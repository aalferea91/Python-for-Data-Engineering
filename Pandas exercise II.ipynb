1. Knowing that the data is from 2019, load the file 'fifa19.csv' and create a new column called 'birth' that show the birth year of each player.

2. Update the column `Release Clause` to show the currency value in millions.

3. Create a column named `effort` which will be the difference between the column `Potential` and the column `Overall`. Which players are in the top 10 according to this field?

4. Update the column `Height` to return the value in cm. Currently the measure is in feet, the conversion rate is: `1 feet = 30.48cm`.

```python
import pandas as pd
```

```python
pd.read_csv('data/fifa19.csv').head()
```

```python
df_fifa=(
    pd.read_csv('data/fifa19.csv')
    .assign(
        birth=lambda x: 2019-x['Age'])
)
```

```python
df_fifa.head()
```

```python
#observe that some values are in K and some in M
df_fifa['Value']
```

```python
#we create a function that will transform the strings into a float
def transform(x):
    if x[-1]=='M':
        return float(x[1:-1])
    elif x[-1]=='K':
        return float(x[1:-1])/1000
```

```python
df_fifa['Release Clause']=df_fifa['Release Clause'].map(lambda x: transform(x))
```

```python
df_fifa.head()
```

```python
df_fifa=(df_fifa
    .assign(
        effort=lambda x: x['Potential']-x['Overall'])
    .sort_values(by='effort',axis=0,ascending=False)
)
```

```python
df_fifa.head(10)
```

```python
def feet2cm(x):
    feet,inches=x.split("'")
    cm=float(feet)*30.48+float(inches)*2.54
    return cm
```

```python
df_fifa['Height']=df_fifa['Height'].map(feet2cm)
```

```python
df_fifa.head()
```

The dataset `baby-names` contains information about the names of the newborn babies in the US over different years. This dataset comes in different files, one file per year and each year is specified in the name of the file.

Create a function that accepts a list of paths and creates a dataframe with all the data combined with the information of the year into a new column.

```python
def read_babyname_files(paths):
    return pd.concat([pd.read_csv(path).assign(year=path.split('/')[-1][:4]) for path in paths])
```

```python
paths=["./data/2014-baby-names-raw.csv","./data/2015-baby-names-raw.csv"]
```

```python
read_babyname_files(paths)
```
