#### 5. Load each dataset in a different dataframe:

```python
df_confirmed = pd.read_csv('data/covid_confirmed.csv')
df_deaths = pd.read_csv('data/covid_deaths.csv')
df_recovered = pd.read_csv('data/covid_recovered.csv')
```

#### 6. Combine the 3 dataframes into one joining them by the fields `Date` and `Country`.

```python
df_spain_by_day=(
    df_confirmed
    .merge(df_deaths,on=['Date','Country'])
    .merge(df_recovered,on=['Date','Country'])
    [['Country','Date','Confirmed','Deaths','Recovered']]
)
```

```python
df_spain_by_day.head()
```

#### 7. Divide the field `Date` in 3 different fields `Year`, `Month` y `Day`.

```python
df_spain_by_day.head(3)
```

```python
df_spain_by_day['Date'][0]
```

```python
(
    df_spain_by_day
    .assign(
        Year=lambda df:df['Date'].apply(lambda x:int(x[:4])),
        Month=lambda df:df['Date'].apply(lambda x:int(x[5:7])),
        Day=lambda df:df['Date'].apply(lambda x:int(x[8:10])))
)
```

```python
#another way to do this is by transforming the column date into a series of datetime records
df_spain_by_day.Date = pd.to_datetime(df_spain_by_day.Date)
```

```python
df_spain_by_day=(
    df_spain_by_day
    .assign(
        Year=lambda df: df.Date.dt.year,
        Month=lambda df: df.Date.dt.month,
        Day=lambda df: df.Date.dt.day
    )
)
```

```python
df_spain_by_day.head()
```

#### 8. Calculate the mortality ratio per month. Which month is the one with the highest mortality?

```python
# Completar
```

```python
(
    df_spain_by_day
    .assign(
        mortality=lambda df: df['Deaths']/df['Confirmed']*100
    )
    .groupby('Month')
    .mean()
    .nlargest(3,'mortality')
)
```

#### 9. Calculate the total deaths that ocurred each month. Which month has the highest number of deaths and which month has the lowest?

```python
def death_month(deaths):
    return deaths.max() - deaths.min()

(
    df_spain_by_day
    .assign(
        mortality=lambda df: df.Deaths/df.Confirmed*100
    )
    .groupby('Month')
    .agg(
        {
            "Deaths": death_month
        }
    )
)
```
